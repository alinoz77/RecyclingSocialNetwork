```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
users_df = pd.read_json('./users_export.json')
```

```python
users_by_country = users_df['country'].value_counts()
```

```python
users_by_country[users_by_country > 10].plot(kind='barh')
```

```python
users_by_quarter = users_df['timeinterval'].value_counts()
```

```python
users_by_quarter.plot(kind='barh')
```
